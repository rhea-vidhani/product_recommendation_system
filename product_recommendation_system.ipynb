# ***Collaborative filtering***

```python
#Create user-item interaction matrix
user_item_matrix = df.pivot_table(
    index='user_id',
    columns='product_id',
    values='rating',
    fill_value=0
)
```

```python
#Compute Item-Item Cosine Similarity
from sklearn.metrics.pairwise import cosine_similarity
item_similarity = cosine_similarity(user_item_matrix.T)
item_similarity_df = pd.DataFrame(item_similarity,
                                  index=user_item_matrix.columns,
                                  columns=user_item_matrix.columns)
```

# ***Content - Based filtering***

```python
# Deduplicate before TF-IDF
prod = df[['product_id', 'product_name', 'category', 'about_product']].drop_duplicates(subset=['product_id']).copy()
```

```python
#Combine Features for Text Vectorization
for col in ['product_name', 'category', 'about_product']:
    prod[col] = prod[col].astype(str).fillna("")

prod['text_features'] = (
    prod['product_name'].str.lower() + " " +
    prod['category'].str.lower() + " " +
    prod['about_product'].str.lower()
)
```

```python
#Convert Text to TF-IDF Vectors
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(prod['text_features'])
```

```python
#Compute Item-Item Content Similarity
content_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
content_sim_df = pd.DataFrame(content_sim,
                              index=prod['product_id'],
                              columns=prod['product_id'])
```

# ***Hybrid filtering***

```python
# Ensure product IDs in both similarity matrices are aligned
common_products = sorted(set(item_similarity_df.index) &
                       set(content_sim_df.index))

# Slice both to common products
item_sim_common = item_similarity_df.reindex(index=common_products, columns=common_products)
content_sim_common = content_sim_df.reindex(index=common_products, columns=common_products)

# Compute hybrid similarity
hybrid_sim_df = 0.7 * item_sim_common + 0.3 * content_sim_common
```

# ***Recommendation System***

```python
# Ask user to input a product ID
sample_product_id = input("Enter a product_id to get recommendations: ").strip()
print(f"You entered: {sample_product_id}")

# Safety check: make sure product_id exists
if sample_product_id not in hybrid_sim_df.index:
    print("Product not found in hybrid similarity matrix.")
else:
    # Filter similarity scores for the selected product
    filtered_scores = hybrid_sim_df.loc[sample_product_id]

    # Drop self-score (similarity with itself = 1.0 usually)
    filtered_scores = filtered_scores.drop(sample_product_id, errors='ignore')

    # Get strong matches above threshold
    strong_matches = filtered_scores[filtered_scores > 0.2].sort_values(ascending=False)
    print("\nTop similar products (sim > 0.2):")
    print(strong_matches.head())

    # Top 5 regardless of threshold
    top_ids = filtered_scores.nlargest(5).index
    recommended_products = df[df['product_id'].isin(top_ids)][['product_id', 'product_name', 'category']]
    print("\nRecommended product details:")
    print(recommended_products.reset_index(drop=True))
```

# ***Model Evaluation***

```python
def evaluate_hybrid_recommender(df, hybrid_sim_df, k=5, min_user_interactions=2):
    hits, precision_sum, recall_sum, evaluated_products = 0, 0, 0, 0

    for product in df['product_id'].unique():
        if product not in hybrid_sim_df.index:
            continue

        # Users who interacted with this product
        users = df[df['product_id'] == product]['user_id'].unique()
        related_products = df[(df['user_id'].isin(users)) &
                              (df['product_id'] != product)]['product_id'].unique()
        if len(related_products) < min_user_interactions:
            continue

        # Always get a full row (Series) of length
        scores = hybrid_sim_df.loc[product]

        # Top-K recommendations
        top_recs = scores.sort_values(ascending=False).head(k).index

        # Metrics
        hits += any(item in related_products for item in top_recs)
        precision_sum += sum(1 for item in top_recs if item in related_products) / k
        recall_sum += sum(1 for item in top_recs if item in related_products) / len(related_products)

        evaluated_products += 1

    return {
        "HitRate@K": hits / evaluated_products if evaluated_products else 0,
        "Precision@K": precision_sum / evaluated_products if evaluated_products else 0,
        "Recall@K": recall_sum / evaluated_products if evaluated_products else 0,
        "Evaluated Products": evaluated_products
    }
```

```python
metrics = evaluate_hybrid_recommender(df, hybrid_sim_df, k=5)
print("Evaluation Results for Hybrid Recommender:")
print(metrics)
```
