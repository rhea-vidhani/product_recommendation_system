# ***Data Cleaning***

```python
#Identify missing values
df.isnull().sum()
```

```python
#Remove rows which have missing values
df.dropna(inplace=True)
```

```python
#Change datatype
df['discounted_price'] = df['discounted_price'].str.replace('₹', '', regex=False).str.replace(',', '', regex=False).astype(float)
df['actual_price'] = df['actual_price'].str.replace('₹', '', regex=False).str.replace(',', '', regex=False).astype(float)
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '', regex=False).astype(float)
df['rating_count'] = df['rating_count'].str.replace(',', '', regex=False).astype(float)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
#Create variables
df['purchase_price'] = df['discounted_price']
df['rating_weight'] = df['rating'] * df['rating_count']
```

```python
#Dropping columns
columns_to_drop = ['discounted_price','discount_percentage', 'user_name', 'review_id', 'review_title',
                           'review_content', 'product_link', 'img_link', 'rating', 'rating_count']
df.drop(columns=columns_to_drop, axis=1, inplace=True, errors='ignore')
```

```python
#Removing duplicate product entries by 'product_id'
df.drop_duplicates(subset='product_id', keep='first', inplace=True)
```

```python
#Remove rows which has missing values
df.dropna(inplace=True)
```

```python
#Cleaned dataset
df.info()
```

# ***Collaborative filtering***

```python
#Create user-item interaction matrix
user_item_matrix = df.pivot_table(
    index='user_id',
    columns='product_id',
    aggfunc='size',
    fill_value=0
)
```

```python
#Compute Item-Item Cosine Similarity
from sklearn.metrics.pairwise import cosine_similarity
item_similarity = cosine_similarity(user_item_matrix.T)
item_similarity_df = pd.DataFrame(item_similarity,
                                  index=user_item_matrix.columns,
                                  columns=user_item_matrix.columns)
```

# ***Content - Based filtering***

```python
#Combine Features for Text Vectorization
df['text_features'] = df['product_name'] + ' ' + df['category'] + ' ' + df['about_product']
```

```python
#Convert Text to TF-IDF Vectors
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(df['text_features'])
```

```python
#Compute Item-Item Content Similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
content_sim_df = pd.DataFrame(cosine_sim, index=df['product_id'], columns=df['product_id'])
```

# ***Hybrid filtering***

```python
# Ensure product IDs in both similarity matrices are aligned
common_products = list(set(item_similarity_df.index) &
                       set(content_sim_df.index))

# Slice both to common products
item_sim_common = item_similarity_df.loc[common_products, common_products]
content_sim_common = content_sim_df.loc[common_products, common_products]

# Compute hybrid similarity
hybrid_sim_df = (0.7 * item_sim_common) + (0.3 * content_sim_common)
```

# ***Recommendation System***

```python
# Ask user to input a product ID
sample_product_id = input("Enter a product_id to get recommendations: ")
print(f"You entered: {sample_product_id}")

# Filter similarity scores for the selected product
filtered_scores = hybrid_sim_df.loc[sample_product_id]
strong_matches = filtered_scores[filtered_scores > 0.7].sort_values(ascending=False)
print("Top similar products:")
print(strong_matches.head(5))

# Map top product IDs to details
top_ids = filtered_scores.nlargest(5).index
recommended_products = df[df['product_id'].isin(top_ids)][['product_id', 'product_name', 'category']]
print(recommended_products)
```

# ***Model Evaluation***

```python
#Display hybrid similarity scores
import matplotlib.pyplot as plt

plt.hist(hybrid_sim_df.values.flatten(), bins=50)
plt.title('Distribution of Hybrid Similarity Scores')
plt.xlabel('Similarity Score')
plt.ylabel('Frequency')
plt.show()
```

```python
def evaluate_similarity_recommender(df, hybrid_sim_df, k=5, min_user_interactions=2):
    hits, precision_sum, recall_sum, evaluated_products = 0, 0, 0, 0

    product_ids = df['product_id'].unique()

    for product in product_ids:
        users_who_bought = df[df['product_id'] == product]['user_id'].unique()

        # Get all other products these users interacted with (excluding the input product)
        related_products = df[df['user_id'].isin(users_who_bought) & (df['product_id'] != product)]['product_id'].unique()
        if len(related_products) == 0:
            continue

        if product not in hybrid_sim_df.index:
            continue

        top_recs = hybrid_sim_df.loc[product].sort_values(ascending=False).head(k).index

        hits += any(item in related_products for item in top_recs)
        recall_sum += sum(1 for item in top_recs if item in related_products) / len(related_products)
        evaluated_products += 1

    return {
        "HitRate@K": hits / evaluated_products if evaluated_products else 0,
        "Recall@K": recall_sum / evaluated_products if evaluated_products else 0,
        "Evaluated Products": evaluated_products
    }

metrics = evaluate_similarity_recommender(df, hybrid_sim_df, k=5)
print("Evaluation Results for Similarity Recommender:")
print(metrics)
```

```python
import gradio as gr
import joblib

# Load your saved hybrid similarity matrix
hybrid_sim_df = joblib.load('hybrid_sim_df.pkl')

def recommend(product_id, k):
      try:
        if product_id not in hybrid_sim_df.index:
            return f"Product ID '{product_id}' not found in similarity matrix."

        # Get top-K similar product IDs
        top_k = hybrid_sim_df.loc[product_id].sort_values(ascending=False).head(k)
        recommendations = top_k.index.tolist()

        # Create DataFrame and merge with product names from df
        result_df = pd.DataFrame({'product_id': recommendations})
        result_df = result_df.merge(df[['product_id', 'product_name']].drop_duplicates(),
                                    on='product_id', how='left')

        # Format the output with both ID and name
        result_text = "\n".join(
            f"{row['product_id']} — {row['product_name'] if pd.notna(row['product_name']) else 'Name Not Found'}"
            for _, row in result_df.iterrows()
        )

        return result_text

      except Exception as e:
        return f"Error: {str(e)}"
```

```python
import gradio as gr

interface = gr.Interface(
    fn=recommend,
    inputs=[
        gr.Textbox(label="Enter Product ID"),
        gr.Slider(minimum=1, maximum=10, step=1, value=5, label="Number of Recommendations")
    ],
    outputs=gr.Textbox(label="Top Similar Products"),
    title="Product Recommendation System",
    description="Get top similar products with names."
)

interface.launch()
```
